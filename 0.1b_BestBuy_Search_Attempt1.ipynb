```python
# Dependencies
import requests
import json
import csv
import pandas as pd
import time
import re
import matplotlib.pyplot as plt

def print_r(text):
   print(json.dumps(text, indent=4, sort_keys=False))
```

```python
url = "https://api.bestbuy.com/v1/products((categoryPath.id=abcat0502000))?apiKey=Jz4AJitfYbGIGGaOuTR09Lpl&sort=customerReviewAverage.asc&show=customerReviewAverage,name,customerReviewCount,inStoreAvailability,onlineAvailability,regularPrice,salePrice,upc,manufacturer&pageSize=100&page="
```

```python
laptop_list = []


for page in range(1, 9):

    # create query with page number
    query_url = url + str(page) + "&format=json"

    products = requests.get(query_url).json()
    print(query_url)
    # Add a one second interval between queries to stay within API query limits
    time.sleep(1)
    
    # loop through the response and append each article to the list
    for product in products["products"]:
        
        try:
            product["screensize"] = re.findall(r'[0-9\.]+"', product["name"])[0].replace("\"","")
        except (KeyError, IndexError):
            print_r(product)

        laptop_list.append(product)
```

```python
laptopbestbuy_df = pd.DataFrame(laptop_list)
laptopbestbuy_df


laptopbestbuy_df['screensize'] = pd.to_numeric(laptopbestbuy_df['screensize'])
laptopbestbuy_df.dtypes
laptopbestbuy_df
```

```python
laptopbestbuy_df.to_csv("LaptopBestbuy.csv", index=False, header=True)
```

```python
bins = [0, 14, 16,20]
group_names = ["14 inches or Less", "14 to 16 inches", "Greater than 16 inches"]
laptopbestbuy_df["Screen Size Category"] = pd.cut(laptopbestbuy_df["screensize"], bins, labels=group_names)
laptopbestbuy_df
```

```python
laptopbestbuy_df

sizegroup = laptopbestbuy_df["Screen Size Category"].value_counts()
sizegroup_df = pd.DataFrame(sizegroup)
sizegroup_df.head()
```

```python
grouped_df = laptopbestbuy_df.groupby(['Screen Size Category'])
print(grouped_df)
grouped_df.head()
```

```python

averageprice = grouped_df["regularPrice"].mean()
averagescore = grouped_df["customerReviewAverage"].mean()
```
